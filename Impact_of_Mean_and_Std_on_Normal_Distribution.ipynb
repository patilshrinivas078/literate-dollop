# Impact of Mean and Standard Deviation on Normal Distribution Patterns

The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution that is symmetric about its mean, showing that data near the mean are more frequent in occurrence than data far from the mean. The shape and position of the normal distribution are determined by its mean ($\mu$) and standard deviation ($\sigma$).

**Key Characteristics of the Normal Distribution**

**Mean ($\mu$):**
- The mean is the central value of the distribution. It determines the location of the center of the graph.
- The entire distribution is centered around the mean, meaning that the peak of the bell curve occurs at $\mu$.
- Changing the mean shifts the distribution left or right on the horizontal axis without altering its shape.

**Standard Deviation ($\sigma$):**
- The standard deviation measures the dispersion or spread of the distribution. It determines the width of the bell curve.
- A smaller standard deviation results in a narrower and taller curve, indicating that the data points are closer to the mean.
- A larger standard deviation results in a wider and flatter curve, indicating that the data points are more spread out from the mean.

**Mathematical Representation**

The probability density function (pdf) of the normal distribution is given by:

$$
f(x \mid \mu, \sigma) = \frac{1}{\sqrt{2 \pi \sigma^2}} e^{-\frac{(x - \mu)^2}{2 \sigma^2}}
$$

where:
- $x$ is a variable.
- $\mu$ is the mean.
- $\sigma$ is the standard deviation.

**Visual Explanation**

**Effect of Mean ($\mu$):**
- If $\mu = 0$, the distribution is centered at zero.
- If $\mu$ is positive, the distribution shifts to the right.
- If $\mu$ is negative, the distribution shifts to the left.

For example:
- For $\mu = 0$, the peak is at $x = 0$.
- For $\mu = 2$, the peak is at $x = 2$.
- For $\mu = -2$, the peak is at $x = -2$.

**Effect of Standard Deviation ($\sigma$):**
- If $\sigma = 1$, the distribution has a standard spread.
- If $\sigma < 1$, the distribution is more peaked (narrower).
- If $\sigma > 1$, the distribution is more spread out (wider).

For example:
- For $\sigma = 1$, the curve is moderately spread.
- For $\sigma = 0.5$, the curve is narrower and taller.
- For $\sigma = 2$, the curve is wider and shorter.


```python
import numpy as np
import matplotlib.pyplot as plt
```

**The following function plots a Normal Distribution**

```python
def plot_normal_distribution(mu, sigma):
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
    y = (1 / (np.sqrt(2 * np.pi * sigma**2))) * np.exp(-(x - mu)**2 / (2 * sigma**2))
    plt.plot(x, y, label=f'µ={mu}, σ={sigma}')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.title('Normal Distribution')
    plt.legend()
```

**Now, let's try to plot normal distributions with different means and standard deviations**

$$
\begin{aligned}
1. \quad & \mu = 0, \sigma = 1 \\
2. \quad & \mu = 2, \sigma = 1 \\
3. \quad & \mu = 0, \sigma = 0.5 \\
4. \quad & \mu = 0, \sigma = 2
\end{aligned}
$$


```python
plt.figure(figsize=(10, 6))
plot_normal_distribution(0, 1)   # µ=0, σ=1
plot_normal_distribution(2, 1)   # µ=2, σ=1
plot_normal_distribution(0, 0.5) # µ=0, σ=0.5
plot_normal_distribution(0, 2)   # µ=0, σ=2
plt.show()
```

The plotted normal distributions illustrate the impact of mean ($\mu$) and standard deviation ($\sigma$) on the shape and position of the curves. The standard normal distribution ($\mu = 0, \sigma = 1$) is centered at 0 with a standard deviation of 1, exhibiting symmetry around the mean. Increasing the mean to 2 ($\mu = 2, \sigma = 1$) shifts the distribution to the right without altering its shape. Reducing the standard deviation to 0.5 ($\mu = 0, \sigma = 0.5$) results in a narrower and taller curve, indicating greater data concentration around the mean. Similarly, increasing the standard deviation to 2 ($\mu = 0, \sigma = 2$) widens the curve and spreads out the data points from the mean. These variations demonstrate how changes in mean and standard deviation influence the centrality and spread of normal distributions, offering insights into their statistical characteristics.
